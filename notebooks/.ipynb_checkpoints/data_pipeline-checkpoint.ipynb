```python
badValues = ["N/A", "None", "", " ", "-", "'"]
for df in [productPropertiesDf, productDescriptionsDf, manufacturersDf]:
    df.replace(badValues, np.nan, inplace=True)
```

```python
# After joining product_properties with product_descriptions
mergedDf = productPropertiesDf.merge(productDescriptionsDf, on='Articlenumber', how='left')

# Count rows where 'Short description' is missing (NaN)
missingDescriptionsCount = mergedDf['Short description'].isna().sum()

print(f"Products without a description: {missingDescriptionsCount}")
```

```python
mergedDf
```

```python
# Step 2: Merge with manufacturersdf
finalDf = mergedDf.merge(manufacturersDf, on='Manufacturernumber', how='left')

# First count initial size
initialSize = productPropertiesDf.shape[0]  # This is before any merge
finalSize = finalDf.shape[0]  # This is after merging all datasets

# Calculate rows dropped due to missing values in the merge process
rowsDropped = finalSize - initialSize
print(f"Rows dropped during the merge process: {rowsDropped}")
```
